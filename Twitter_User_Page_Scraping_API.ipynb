## Twitter User's Profile Page Scraping API

**This note book describes how to get list of followers from a specific Twitter User's Profile Page**

The Twython 3.4 python package is used to retrieve the list of followers SCREEN NAME for a specific Twitter Account. The process involves the following steps:
1. Import the required modules (all functions in twython and time for scheduling the streaming)
2. For authentication, insert string values for the keys generated from the Twitter Streamming Application [![](https://abs.twimg.com/a/1404172626/images/oauth_application.png)](https://apps.twitter.com/)
3. Define Twython authentication command with the created keys
4. Use "get_followers_list" function from twython to retrieve list of disctionary of each followers from the specified user's tweet account screen name
5. Run "followers_list" to print the list of followers prefixed with "@"
